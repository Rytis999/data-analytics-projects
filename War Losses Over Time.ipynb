```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('russia_losses_equipment.csv')
df["date"] = pd.to_datetime(df["date"])
```

```python
df.head()




```

```python
df.info()
```

```python
df.fillna(0, inplace=True)

df.head()





```

```python
loss_totals = df.select_dtypes(include='number').drop(columns="day")
loss_sums = loss_totals.sum().sort_values(ascending=False)
loss_sums.head(10).plot(kind="barh", title="TOP 10 technikos praradim tip")
plt.xlabel("Bendra suma")
plt.tight_layout()
plt.show()

```

```python
cols_to_plot = ["tank", "drone", "aircraft", "APC"]
df.set_index("date")[cols_to_plot].plot(figsize=(12,6), title="Praradim dinamika per laik")
plt.ylabel("Praradim skaiius")
plt.tight_layout()
plt.show()
```

```python
# 5. Mnesin dinamika
df_monthly = df.set_index("date").resample("M").sum()
df_monthly[cols_to_plot].plot(title="Mnesin technikos praradim dinamika")
plt.tight_layout()
plt.show()
```

```python
# 6. Daniausiai minimos kryptys
if "greatest losses direction" in df.columns:
    direction_counts = df["greatest losses direction"].value_counts().head(10)
    direction_counts.plot(kind="bar", title="TOP 10 praradim krypi")
    plt.ylabel("Kiek kart paminta")
    plt.tight_layout()
    plt.show()
```

```python
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Koreliacij matrica tarp technikos tip")
plt.tight_layout()
plt.show()
```

```python
# 8. Dienos su didiausiais praradimais
df["total_loss_per_day"] = df.select_dtypes(include='number').sum(axis=1)
top_days = df.sort_values("total_loss_per_day", ascending=False)[["date", "total_loss_per_day"]].head(10)
print("TOP 10 dien pagal bendrus technikos praradimus:")
print(top_days)
```
